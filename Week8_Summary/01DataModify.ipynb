```python
import gdown
import pandas as pd

# 아래 두개의 값 입력 (로컬 컴퓨터 환경에서 실행은 url 비워둘 것)
filename = 'PtLibraryData.csv'  # 예제에서 사용할 filename 입력
url = ''    # 여기에 본인이 가진 링크의 난수 입력

if url != '' :
    gdown.download(f'https://drive.google.com/uc?id={url}', filename, quiet=False)
    dir = ''
else :
    dir = '../_csv/'  # 로컬 컴퓨터 환경에서 사용할 dir (예제는 _csv)

pt_book = pd.read_csv(f'{dir}{filename}', low_memory=False)
pt_book.head()
```

```python
pt_book.isna().sum()    # NaN으로 되어있는 항목들의 합을 구함
```

```python
pt_book.loc[0,'도서권수'] = None
pt_book['도서권수'].isna().sum()
```

```python
pt_book.head(2)
```

```python
pt_book.loc[0,'도서권수'] = 1       # 도서권수 index 0번 1로 변경
pt_book = pt_book.astype({'도서권수':'int32', '대출건수': 'int32'})     # 파이썬 딕셔너리 int32로 변경
pt_book.head(2)
```

```python
set_isbn_na_rows = pt_book['세트 ISBN'].isna()  # 누락된 값 불리언 배열로

pt_book.loc[set_isbn_na_rows, '세트 ISBN'] = ''     # 누락된 위치 공백으로
pt_book.head()
```

```python
import numpy as np

# pt_book.fillna('없음')    # NaN -> 없음 으로 바꾸기
pt_tmp = pt_book.replace(np.nan, '없음')    # NaN -> 없음 으로 바꾸기
```

```python
pt_tmp.head()
```

```python
pt_book.replace([np.nan, '2023'], ['없음', '23']).head(2)   # 리스트를 이용한 값 수정
pt_book.replace({np.nan :'없음'}, {'2023': '23'}).head(2)   # 딕셔너리를 이용한 값 수정
pt_book.replace({'부가기호': np.nan}, '없음').head(2)   # 특정 항목의 값 수정
```

```python
pt_book['발행년도'].str.contains('1988').sum()  # 발행년도에 문자열 '1988'이 포함되어있는 갯수
```

```python
# 숫자이외의 발행년도 확인
invalid_number = pt_book['발행년도'].str.contains('\D', na = True)
print(invalid_number.sum())
pt_book[invalid_number].head()
```

```python
pt_book2 = pt_book.replace({'발행년도':'.*(\d{4}).*'}, r'\1', regex=True)       # 발행년도 숫자 4자리 앞 뒤에 있는 문자, 숫자를 제거하는 정규표현식 사용
pt_book2[invalid_number].head()
```

```python
# 숫자 이외의 발행년도 확인
unknown_year = pt_book2['발행년도'].str.contains('\D', na = True)   # 위에서 값을 수정해도 발행년도를 확인 할 수 없는 시리즈들을 저장
print(unknown_year.sum())
pt_book2[unknown_year].head()
```

```python
pt_book2.loc[unknown_year, '발행년도'] = '-1'       # 값을 확인할 수 없는 발행년도를 -1로 설정한다.
pt_book2 = pt_book2.astype({'발행년도': 'int32'})   # 모든 발행년도가 숫자형이기 때문에 형식을 정수형으로 바꿔줌
```

```python

pt_book2[pt_book2['발행년도'].gt(4000)].head()  # 발행년도 > 4000 일 경우
pt_book2[pt_book2['발행년도'].lt(1900)].head()  # 발행년도 < 1900 일 경우
```
